```python
import matplotlib.pyplot as plt
plt.scatter([0,2,1],[1,1,-1])
plt.show()
```

```python
fig = plt.figure(figsize = (7,7))
plt.scatter([0,2,1], [1,1,-1], s=1000)
plt.show()
```

```python
plt.scatter([range(1,10001)], [sin(x) for x in range(1,10001)], s=1)
plt.xscale('log')
plt.show()
```

```python
plt.scatter([log10(x) for x in range(1,10001)], [sin(x) for x in range(1,10001)], s=1)
plt.show()
```

```python
def f(x):
    if(x<=0):
        return 2**x
    else:
        return 2**(-x)
```

```python
%matplotlib inline
```

```python
plotFunc2D(f, [-20,20])
```

```python
from plot import *
from math import *
plotSurface3D(lambda x,y: (x,(cos(y)*f(x)), (sin(y)*f(x))), [-pi, pi], zRange=[-pi, pi], yRange=[-pi,pi])
```

```python
plotSurface3D(lambda u, v: ((4 + cos(v)) * cos(u),
                              (4 + cos(v)) * sin(u),
                              sin(v)),
                [0, 2 * pi], zRange=[-pi, pi])
```
